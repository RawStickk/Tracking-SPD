```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def plot_graph(pipes, hits, track):
    
    plt.figure(figsize = (20, 20))
    
    plt.grid()
   
    plt.xlabel(r'X', fontsize = 14)
    plt.ylabel(r'Y', fontsize = 14)
    plt.title(r'Detector pipes and hits', fontsize = 16)
    
    plt.xticks(np.arange(-3000, 3000, 200))
    plt.yticks(np.arange(-3000, 3000, 200))
    
    parallel_pipes = pipes[pipes["alpha"] == 0]
    inclined_pipes = pipes[pipes["alpha"] == 5]
    
    plt.errorbar(inclined_pipes["x"], inclined_pipes["y"], fmt = ".")
    plt.errorbar(parallel_pipes["x"], parallel_pipes["y"], fmt = ".")
    #plt.errorbar(track["x"], track["y"], fmt = ".")
    for i in range(1, hits["NTRK"].max() + 1):
        df = hits[hits["NTRK"] == i]
        #display(df)
        plt.errorbar(df["x"], df["y"], fmt = ".")
    
```

```python
pipes = pd.read_csv("wires.csv")
hits = pd.read_csv("hits.csv")
track = pd.read_csv("track.csv")
#display(hits[hits["NTRK"] == 7])
```

```python
plot_graph(pipes, hits, track)
```
